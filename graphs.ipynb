```python
import pandas as pd

# Data provided
people = ['Kiran', 'Arun', 'Vijay', 'Varun']
age = [25, 30, 35, 40]
height = [145, 151, 165, 173]  # in cm
weight = [45, 55, 65, 75]      # in kg

# Creating a DataFrame
health_data = pd.DataFrame({
    'Name': people,
    'Age': age,
    'Height(cm)': height,
    'Weight(kg)': weight
})

# Display the dataset
print(health_data)
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

# Scatter Plot - Height vs Weight
plt.figure(figsize=(8, 6))
plt.scatter(health_data['Height(cm)'], health_data['Weight(kg)'], color='blue')
plt.title('Scatter Plot: Height vs Weight')
plt.xlabel('Height (cm)')
plt.ylabel('Weight (kg)')
plt.grid(True)
plt.show()

# Bar Chart - Age of Individuals
plt.figure(figsize=(8, 6))
plt.bar(health_data['Name'], health_data['Age'], color='green')
plt.title('Bar Chart: Age of Individuals')
plt.xlabel('Name')
plt.ylabel('Age (years)')
plt.show()

# Histogram - Distribution of Height
plt.figure(figsize=(8, 6))
plt.hist(health_data['Height(cm)'], bins=5, color='orange', edgecolor='black')
plt.title('Histogram: Distribution of Height')
plt.xlabel('Height (cm)')
plt.ylabel('Frequency')
plt.show()

# Histogram - Distribution of Weight
plt.figure(figsize=(8, 6))
plt.hist(health_data['Weight(kg)'], bins=5, color='purple', edgecolor='black')
plt.title('Histogram: Distribution of Weight')
plt.xlabel('Weight (kg)')
plt.ylabel('Frequency')
plt.show()
```
